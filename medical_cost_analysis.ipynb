```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('insurance.csv')
```

We must convert this categorical data into nominal data for us to evaluate it, we do that with the LabelEncoder class of the sklearn.preprocessing package

```python
le = LabelEncoder()
le.fit(df.sex.drop_duplicates())
df.sex = le.transform(df.sex)

le.fit(df.smoker.drop_duplicates())
df.smoker = le.transform(df.smoker)

le.fit(df.region.drop_duplicates())
df.region = le.transform(df.region)
```

```python
corr = df.corr()
df.head()
```

```python
sns.heatmap(corr, annot=True, cbar=True)
plt.show()
```

Let's see the distribution of the medical cost

```python
df.charges.plot(kind='hist')
plt.xlabel('Amount in $')
plt.legend()
plt.show()
```

The distribution clearly shows that the majority of people fall within the 1000 dollars to 13000 dollars range approximately. Let us check the cost distribution for people who are aged 18 and aged above 18.

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(121)
ax = sns.distplot(df['charges'][df['age'] == 18])
plt.title('Age == 18')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('Cost')

ax = fig.add_subplot(122)
ax = sns.distplot(df['charges'][df['age'] > 18])
plt.title('Age > 18')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('Cost')


plt.show()
```

Let us see how many people smoke in this dataset, and how many people don't:

```python
df.smoker.value_counts()
```

```python
plt.xlabel('Non-Smoker = 0, Smoker = 1')
plt.ylabel('Frequency')
plt.title('Smoker Distribution')
df.smoker.value_counts().plot(kind='bar')
plt.show()
```

There are about 1/3rd amount of smokers in this dataset compared to the non-smokers. Let us see how does the cost vary for a smoker and a non smoker.

```python
sns.catplot(x='smoker', y='charges', data=df)
plt.show()
```

It was always obvious that a smoker will have to spend more in terms of medical costs, however, the difference is significantly visible

Let us see how many people actually smoke at the age of 18 and the medical cost they face:

```python
sns.catplot(x='smoker', y='charges', hue='sex', kind='violin', data=df[df['age'] == 18], legend=True)
plt.show()
```

This clearly shows that the people start spending a lot on medical resources even at the age of 18 if they smoke, compared to those who don't, irrespective of their sex.

Let us see the same result for people older than 18:

```python
sns.catplot(x='smoker', y='charges', hue='sex', kind='violin', data=df[df['age'] > 18], legend=True)
plt.show()
```

Still, no significant difference at all, people who smoke will always tend to spend more on medical facilities

Let us see how the BMI varies between different people:

```python
sns.distplot(df['bmi'])
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.legend()
plt.show()
```

We can see that the average BMI in this dataset is approximately 30, which is pretty bad. BMI categories can be divided into 4:

1. Underweight - BMI < 18.5
2. Normal - BMI > 18.5 and BMI < 24.9
3. Overweight - BMI > 24.9 and BMI < 29.9
4. Obese - BMI > 30

Let's see how medical costs are distributed amongst these categories:

```python
fig = plt.figure(figsize=(20,20))

ax = fig.add_subplot(221)
ax = sns.distplot(df['charges'][df['bmi'] < 18.5], rug=True)
plt.title('Underweight (BMI <18.5)')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('BMI')


ax = fig.add_subplot(222)
ax = sns.distplot(df['charges'][(df['bmi'] >= 18.5) & (df['bmi'] < 24.9)], rug=True)
plt.title('Normal (BMI > 18.5 and BMI < 24.9)')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('Cost')


ax = fig.add_subplot(223)
ax = sns.distplot(df['charges'][(df['bmi'] >= 24.9) & (df['bmi'] < 29.9)], rug=True)
plt.title('Overweight (BMI > 24.9 and BMI < 29.9)')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('Cost')


ax = fig.add_subplot(224)
ax = sns.distplot(df['charges'][df['bmi'] > 29.9], rug=True)
plt.title('Obese (BMI > 29.9)')
plt.xticks([0,10000,20000,30000,40000,50000],np.arange(0,50000,10000))
plt.xlabel('Cost')


plt.show()
```

The cost marginally increases as we proceed with the BMI categories. Let us see how the cost varies with BMI when the person smokes or does not smoke:

Let's see how having children affects smoking habits:

```python
sns.catplot(x='children', y='charges', hue='smoker', kind='box', data=df[df['children'] == 0])
plt.show()
```

```python
sns.catplot(x='children', y='charges', hue='smoker', kind='box', data=df[df['children'] > 0])

plt.show()
```

```python
fig = plt.figure(figsize=(20,7))

ax=fig.add_subplot(121)
ax = df[df['children'] == 0].smoker.value_counts().plot(kind='bar', color='red')
plt.title('No Children')
plt.xlabel('Smoker')
plt.ylabel('Frequency')

ax=fig.add_subplot(122)
df[df['children'] > 0].smoker.value_counts().plot(kind='bar', color='red')
plt.title('Atleast 1 Child')
plt.xlabel('Smoker')
plt.ylabel('Frequency')

plt.show()
```

It is clearly visible that majority of the people who had atleast 1 children consisted of less smokers than those who did not have any children

Let us have a similar plot that tells us about the frequency of number of children:

```python
fig = plt.figure(figsize=(20,7))

ax=fig.add_subplot(121)
ax = df[df['smoker'] == 0].children.value_counts().plot(kind='bar', color='green')
plt.title('Non-Smoker')
plt.xlabel('Children')
plt.ylabel('Frequency')

ax=fig.add_subplot(122)
ax = df[df['smoker'] == 1].children.value_counts().plot(kind='bar', color='green')
plt.title('Smoker')
plt.xlabel('Children')
plt.ylabel('Frequency')

plt.show()
```

This clearly indicates that as the no. of children, people intended to leave smoking/didn't smoke because they had children. There is a significantly visible difference.

Let us now see whether the region that the person lives in plays any role in the medical cost:

```python
sns.catplot(x='region', y='charges', hue='smoker', kind='bar', data=df)
plt.show()
```

As expected, the region plays barely any role in affecting a person's medical cost if they are a non-smoker, however we can see a little bit higher expense in region 2 for smokers compared to the other regions.

Hence, we can conclude by saying the the most important factor deciding the medical cost in this dataset is whether a person is smoking or not.
Features like region barely make any difference. The second most important feature in this dataset is the age of the person because a person who
is senile obviously would need much more medical care.
